```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np

data = np.random.exponential(scale=10, size=1000)
print(data[:10])
```

```python
df = pd.DataFrame({'values' : data})
df
```

```python
plt.hist(df, bins=30)
plt.show()
```

```python
plt.boxplot(df)
```

```python
from scipy.stats import zscore
df['z'] = zscore(df['values'])

filtered_df = df[(df['z'].abs() < 3 )| (-3 < df['z'].abs())]

```

```python
plt.boxplot(filtered_df)
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

filtered_data = data[(data >= lower) & (data <= upper)]
```

```python
plt.hist(filtered_data)
```

```python
plt.boxplot(filtered_data)
```
